```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

## cv2.imread(b,g,r)  show apposite layer of rgb format

```python
frame = cv2.imread("thor.jpg")
```

```python
plt.imshow(frame)
```

```python
rgb = cv2.cvtColor(frame , cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(rgb)
```

```python
classifier =cv2.CascadeClassifier("haarcascade_frontalface_alt.xml")
```

```python
faces=classifier.detectMultiScale(frame)
```

```python
faces
```

```python
plt.imshow(faces)
```

```python
face = faces[0]
```

```python
x,y,w,h = face
```

```python
thor = rgb[y:y+h ,x:x+w]
```

```python
plt.imshow(thor)
```

```python
out = cv2.rectangle(rgb,(x,y) , (x+w,y+h),(0,0,255) , 4)
```

```python
plt.imshow(out)
```
