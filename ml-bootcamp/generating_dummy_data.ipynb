```python
import numpy as np
```

```python
x=np.linspace(0,2 * np.pi ,100)
```

```python
y=np.sin(x)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(x,y)
```

```python
plt.plot(y,x)
```

```python
p=np.cos(x)
```

```python
plt.plot(x,p)
```

```python
t=np.tan(x)
```

```python
plt.plot(x,t)
```

```python
theta = np.linspace(0,2 * np.pi ,200)
```

```python
r =10
```

```python
l= r * np.cos(theta)
```

```python
m = r *np.sin(theta)
```

```python
plt.plot(l,m)
```

```python
plt.plot(m,l)
```

## figsize is determining the scale

```python
plt.figure(figsize=(10,10))
plt.plot(l,m)
```

```python
plt.figure(figsize=(4,4))
plt.plot(l,m)
```
